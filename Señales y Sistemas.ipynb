## 2. Conceptos básicos de señales
## \begin{equation}
    a)\;  \int_{-\infty}^{\infty} t^{2}e^{-\sin (t)}\cos (2t)\delta (2t-2\pi )dt
    \end{equation}

```python
%matplotlib inline 
#Sirve para graficar entorno
import sympy as sym
sym.init_printing() #Inicializar
t= sym.symbols('t', real=True)
d = sym.DiracDelta(2*t-2*sym.pi)
f = sym.Function('x')((t**2)*(sym.exp(-sym.sin(t)))*(sym.cos(2*t)))
sym.integrate(d*f, (t, -sym.oo, sym.oo))
```

## \begin{equation}
     b)\; \int_{-\infty}^{\infty} u(t-to)-u(t-n\cdot to)-k\cdot \delta(t-m\cdot to)dt\
    \end{equation}

```python
t,to,n,k= sym.symbols('t to n k', real=True)
sym.integrate(sym.Heaviside(t-to)-sym.Heaviside(t-n*to)-k*sym.DiracDelta(t-n*to), (t, -sym.oo, sym.oo)) 
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = 7
to = 1
x1 = 5*t +1
x2 = (t/2)-2
N = 15
A = 5
B = 3
y1 = (A*x1)-B+(A*x2)-B
y2 = A *(x1+x2)-B
if y1==y2:
    print ('El sistema es lineal')
else: 
        print ('El sistema NO es lineal')
print(y1)
print(y2)

#Se demuestra la invarianza

vt = np.linspace (0,10)
x3 = np.cos(vt)
y3 = A*x3+B
x4 = np.cos(vt-to)
y4= A*x4 +B
plt.plot(y3)
plt.plot(y4)
plt.xlabel(r'$t$')
plt.ylabel(r'$x(t)')
plt.grid()
```

## Hallar la salida y[n] de un SLIT ante la entrada x[n]={-1,1,3,0,5,6,-1}, con una respuesta al impulso h[n]={1,-2,0,1,-1}, donde E{0,-1,+1,-2,+2,...,+-N}

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

s = [0,-3,4,-2,3,1,-1,2,5]
h = [1,-2,0,1,-1,0,0,0,0]
x = [-1,1,3,0,5,6,-1,0,0]
N = int(len(s)/2)
n = np.linspace(-N,N,2*N+1)

plt.stem(n, h, label=r'$h[n]$')
plt.xlim([-N-1,N+1])
plt.xlabel(r'$n$')
plt.ylabel(r'$h[n]$')
plt.grid()
plt.show()

plt.stem(n, x, label=r'$x[n]$')
plt.xlim([-N-1,N+1])
plt.xlabel(r'$n$')
plt.ylabel(r'$x[n]$')
plt.grid()
plt.show()


y = np.convolve(x,h)
ny = np.linspace(-2*N,2*N,4*N+1)
plt.stem(ny, y, label=r'$y[n]$')
plt.xlim([-2*N-1,2*N+1])
plt.xlabel(r'$n$')
plt.ylabel(r'$y[n]$')
plt.grid()
plt.show()

print("h[n]=",h)
print("y[n]=",y)

#s = [0,-3,4,-2,3,1,-1,2,5]
#x = [0,-1,1,-3,0,5,6,-1,0]
```

### Ahora repetimos el proceso pero para  el sistema con respuesta al escalón {-3,4,-2,3,1,2,5}

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

s = [-3,4,-2,3,1,-1,2,5,0]
x = [-1,1,3,0,5,6,-1,0,0]
N = int(len(s)/2)
n = np.linspace(-N,N,2*N+1)

h1 = np.append(0,np.diff(s)) 

plt.stem(n, s, label=r'$s[n]$')
plt.xlim([-N-1,N+1])
plt.xlabel(r'$n$')
plt.ylabel(r'$s[n]$')
plt.grid()
plt.show()

plt.stem(n, h1, label=r'$h[n]$')
plt.xlim([-N-1,N+1])
plt.xlabel(r'$n$')
plt.ylabel(r'$h[n]$')
plt.grid()
plt.show()

plt.stem(n, x, label=r'$x[n]$')
plt.xlim([-N-1,N+1])
plt.xlabel(r'$n$')
plt.ylabel(r'$x[n]$')
plt.grid()
plt.show()


y1 = np.convolve(x,h1)
ny = np.linspace(-2*N,2*N,4*N+1)
plt.stem(ny, y1, label=r'$y1[n]$')
plt.xlim([-2*N-1,2*N+1])
plt.xlabel(r'$n$')
plt.ylabel(r'$y1[n]$')
plt.grid()
plt.show()

print("h1[n]=",h1)
print("y1[n]=",y1)
```
